#Let's do some webscraping! 

```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
#This seems like a good list of big data jargon links
url = "www.datameer.com/company/news/press-releases/"
url2 = "www.bigdatapartnership.com/category/press-releases/"

def getLinks(url):
    
    toReturn = []
    
    r  = requests.get("http://" +url)
    data = r.text
    soup = BeautifulSoup(data)

    for link in soup.find_all('a'):
        l = link.get('href')

        #lots of the links are internal or to anchors on the page. Let's only get full page links
        if (l[0:4] == "http"):
            l = l.replace("http://","").replace("https://","")
            toReturn.append(l)

    return toReturn

bigLinks = getLinks(url) + getLinks(url2)
#print(bigLinks)
```

```python
#Big Data, Big Text
bigText = open("BigDataBigText.txt", "wb")

urls = ["nickstrayer.me", "www.oracle.com/big-data/products.html", "pivotal.io/big-data/press-release/pivotal-introduces-first-open-source-enterprise-grade-big-data-product-suite", \
        "pivotal.io/big-data/press-release/pivotal-changes-the-economics-of-big-data-forever-with-new-pivotal-big-data-suite-offering", \
        "www.talend.com/products/big-data", "blogs.informatica.com/perspectives/2013/07/30/3-big-data-secrets-that-ignite-a-product-launch/"]

for url in bigLinks:

    #url = raw_input("Enter a website to extract the URL's from: ")

    r  = requests.get("http://" +url)

    data = r.text

    soup = BeautifulSoup(data)

    for entry in soup.find_all(['p', 'h1', 'h2']):
        #Get rid of all the darn spaces in there.
        text = ' '.join(entry.text.split()) 

        #take out the new lines
        text = text.rstrip()
        #print(text)

        #Write to the file with a space at the end.
        #In case they use weird encoding put it in a try statement. 
        try: 
            bigText.write(text +" ")
        except: 
            print("dang")
    
#Close da file.    
bigText.flush()
bigText.close()
```
